# Plotting with categorical data

Categorical data is of three types
1. That show observation at the level of each categorical variable.
2. That show abstract representation of each distribution of observations
3. That which apply a statistical functions to show measure of central tendency and confidence

```python
% matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style = 'whitegrid',color_codes = True)

# Globally defining settings for seaborn plots I guess!
```

```python
np.random.seed(sum(map(ord,'categorical')))
```

```python
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
# These are internally loaded inside seaborn because I haven't downloaded any data files for using them.
```

# Categorical Scatter Plots

```python
# Let us see what is contained in tips dataset
tips
```

```python
sns.stripplot(x = 'day',y = 'total_bill', data = tips)
```

observe that the data mapped through stripped plot. It is mapping all individual values for a day.That is strip plot is simply 
stripping it from group and placing it individually.You can also observe that the values are scattered all along the graph.As they are overlapping it is difficult to read values so we can use jitter argument to mark values loosely along a categorical axis.


```python
sns.stripplot(x = 'day',y = 'total_bill',data = tips,jitter = True)
```

One another alternative to use a jitter argument is using a swarm plot.Using an algorithm this swarmplot places the values at differnt locations.

```python
sns.swarmplot(x = 'day',y = 'total_bill',data = tips)
```

Hue parameter allows you to get nested outputs for example,we want to divide the data points we have by sex.This can be done as shown below

```python
sns.swarmplot(x = 'day',y  = 'total_bill',hue = 'sex',data = tips)
```

At the same time note that the differnt colors indicating days is now redundant and color distinguistion is on for sex variable given by hue

```python
sns.swarmplot(y = 'total_bill',x = 'size',data = tips)
```

The order of the variables is selected by itself.

```python
sns.swarmplot(x = 'total_bill',y = 'day',data = tips,hue = 'time')
```

# Distributions of observations within categories

# Box plot

```python
sns.boxplot(x = 'day',y = 'total_bill',data = tips,hue = 'time')
```

As you can see even the box plots we got are nested be time i.e, lunch and dinner.

```python
tips['weekend'] = tips['day'].isin(['sat','sun'])
sns.boxplot(x = 'day',y = 'total_bill',data = tips,hue = 'weekend')
```

because I gave a lower case sat and sun.the weekend series is not created at all or returned false.now lets try the correct column names

```python
tips['weekend'] = tips['day'].isin(['Sat','Sun'])
sns.boxplot(x = 'day',y = 'total_bill',data = tips,hue = 'weekend')
```

# Violin plots

Violin plots is an alternative method in which boxplots and kernel density estimations are combined.

```python
sns.violinplot(x = 'total_bill',y = 'day',hue = 'time',data = tips)
```

How come density estimate can be negative?  But it can be eleminated by specifying scale parameter

```python
sns.violinplot(x = 'total_bill',y = 'day',hue = 'time',data = tips, bw = 0.1,scale = 'count',scale_hue = True)
```

```python
sns.violinplot(x = 'total_bill',y = 'day',hue = 'time',data = tips, bw = 0.1,scale = 'count',scale_hue = False)
```

I didnt really understand how does this scale and scale_hue functions.

```python
sns.violinplot(x = 'day',y = 'total_bill',data = tips,hue = 'sex',split = True)
```

```python
sns.violinplot(x = 'day', y = 'total_bill',data = tips,hue = 'sex',split = True,inner = 'stick',palette = 'Set2')
```

Joining two or more types of graphs is pretty much simple and it doesnot need any command like jointplot or pairplot etc

```python
sns.violinplot(x="day", y="total_bill", data=tips, inner=None)
sns.swarmplot(x="day", y="total_bill", data=tips, color="w",alpha = 0.3);
```

# Drawing Multi-Panel categorical plots.

Multiple plots can be plotted by using factorgrid() which draws a facetgrid()

```python
tips
```

```python
sns.factorplot(x = 'day',y = 'total_bill',hue = 'smoker',data = tips)
```

This will by default print out pointplot but if you need some other graph you can change it by giving kind argument

```python
sns.factorplot(x = 'day',y = 'total_bill',hue = 'smoker',data = tips , kind = 'bar')
```

```python
sns.factorplot(x = 'day',y = 'total_bill',hue = 'smoker',data = tips , col = 'time',kind = 'swarm')
```

```python
sns.factorplot(x = 'time',y = 'total_bill',hue = 'smoker',col = 'day',data = tips, kind = 'violin',size = 3,aspect = 1)
```

```python
g = sns.PairGrid(tips,
             x_vars = ['smoker','sex','weekend'],
             y_vars = ['total_bill','tip'],
             aspect = 1,size = 4)
g.map(sns.violinplot,palette = 'pastel')
```

Pastel is a type of color palette maybe fills colors alternatively
